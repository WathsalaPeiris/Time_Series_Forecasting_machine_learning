```python
# Import our dependencies
import pandas as pd
```

### Data Cleaning Process

Read and Clean CPI Dataset

```python
#  read the Consumer Price Index data

cpi_original_df = pd.read_excel("https://www.abs.gov.au/statistics/economy/price-indexes-and-inflation/consumer-price-index-australia/mar-quarter-2023/640101.xlsx","Data1")
cpi_original_df.head(10)
```

```python
# Get the columns headings
cpi_original_df.keys()
```

```python
#Select columns that contains percentage change from previous period only
selected_cpi_df = cpi_original_df[['Unnamed: 0','Percentage Change from Previous Period ;  All groups CPI ;  Sydney ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Melbourne ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Brisbane ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Adelaide ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Perth ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Hobart ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Darwin ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Canberra ;',
       'Percentage Change from Previous Period ;  All groups CPI ;  Australia ;']]
selected_cpi_df
```

```python
# Drop some rows
selected_cpi_df = selected_cpi_df.drop(labels = range(0,9), axis = 0)
selected_cpi_df
```

```python
# Rename columns
selected_cpi_df.rename(columns = {'Unnamed: 0':'Qtr',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Sydney ;' : 'Sydney',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Melbourne ;' : 'Melbourne',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Brisbane ;' : 'Brisbane',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Adelaide ;' : 'Adelaide',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Perth ;' : 'Perth',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Hobart ;' : 'Hobart',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Darwin ;': 'Darwin',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Canberra ;' : 'Canberra',
                                  'Percentage Change from Previous Period ;  All groups CPI ;  Australia ;' : 'Australia'}, inplace = True)
selected_cpi_df
```

```python
# Change the type of the date column
selected_cpi_df["Qtr"]  = pd.to_datetime(selected_cpi_df.Qtr)
selected_cpi_df["Qtr"].dt.strftime('%d-%m-%Y')
selected_cpi_df
```

```python
#filter data from 1998 onwards
selected_cpi_df = selected_cpi_df[selected_cpi_df['Qtr']>'1997-12-01']
#selected_cpi_df.set_index('Qtr', inplace = True)
selected_cpi_df.head()
```

```python
# save the all data dataframe as a csv file
selected_cpi_df.to_csv("outcome/cpi_all_data.csv", index = False)

#If this file is running in Google colab, uncomment and run the below code.
#selected_cpi_df.to_csv("/content/drive/MyDrive/Data/cpi_all_data.csv", index = False)
```

```python
#filter data from 2009 onwards
final_cpi_df = selected_cpi_df[selected_cpi_df['Qtr']>'2008-12-01']
final_cpi_df.set_index('Qtr', inplace = True)
final_cpi_df.head()
```

```python
# save the dataframe as a csv file
final_cpi_df.to_csv("outcome/cpi_data.csv")

#If this file is running in Google colab, uncomment and run the below code.
#final_cpi_df.to_csv("/content/drive/MyDrive/Data/cpi_data.csv")
```

Read and Clean WPI dataset

```python
wpi_original_df = pd.read_excel("https://www.abs.gov.au/statistics/economy/price-indexes-and-inflation/wage-price-index-australia/dec-2022/634501.xlsx","Data1")
wpi_original_df.head(10)
```

```python
# Get the columns headings
wpi_original_df.keys()
```

```python
# Select only seasonaly adjusted, Private and Public , All Industry column and date
selected_wpi_df = wpi_original_df.loc[: , ['Unnamed: 0', 'Percentage Change From Previous Quarter ;  Australia ;  Total hourly rates of pay excluding bonuses ;  Private and Public ;  All industries ;']]
selected_wpi_df
```

```python
# Delete first few rows
selected_wpi_df = selected_wpi_df.drop(labels = range(0,9), axis = 0)
selected_wpi_df
```

```python
# Rename columns
selected_wpi_df.rename (columns = {'Unnamed: 0' : 'Qtr', 
                                                    'Percentage Change From Previous Quarter ;  Australia ;  Total hourly rates of pay excluding bonuses ;  Private and Public ;  All industries ;' : 'WPI_percentage_change'}, inplace = True )
selected_wpi_df
```

```python
# Change the format of Qtr column
selected_wpi_df["Qtr"] = pd.to_datetime(selected_wpi_df.Qtr)
selected_wpi_df["Qtr"].dt.strftime('%d-%m-%Y')
selected_wpi_df
```

```python
# select data from 1998 onwards
selected_wpi_df = selected_wpi_df[selected_wpi_df['Qtr']>'1997-12-01']
selected_wpi_df
```

```python
# Save the dataftame into a csv file
selected_wpi_df.to_csv("outcome/wpi_all_data.csv", index = False)

#If this file is run in Google colab, uncomment and run below code
#selected_wpi_df.to_csv("/content/drive/MyDrive/Data/wpi_all_data.csv", index = False)
```
